```python
import numpy as np
import pandas as pd
def gbm(n_years=10, n_scenario=1000, mu=0.07,sigma=0.15, steps_per_year=12, s_0=100.0):
    """
    Evolution of a stock price using a stochastic process
    """
    dt = 1/steps_per_year
    n_steps = int(n_years*steps_per_year)
    xi = np.random.normal(size=(n_steps,n_scenario))
    rets = mu*dt+sigma*np.sqrt(dt)*xi
    rets = pd.DataFrame(rets)
    prices = s_0*(1+rets).cumprod()
    return prices
```

```python
p = gbm(10,n_scenario=3)
```
